```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('xclara1.csv')
```

```python

dataset.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(dataset)
kmeans.fit_predict(dataset) 
```

```python

plt.figure(figsize=(12,6))
plt.scatter(dataset['V1'], dataset['V2'], c='b', s=30)
```

```python
plt.scatter(dataset['V1'], dataset['V2'], c=kmeans.labels_)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(dataset['V1'], dataset['V2'], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:,1], marker='*', s=300, c='r')
```
